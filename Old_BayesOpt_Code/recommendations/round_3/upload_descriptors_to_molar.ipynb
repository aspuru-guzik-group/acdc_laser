```python
from MolarInterface import MolarInterface
import pandas as pd
import numpy as np
```

```python
molar = MolarInterface("madness_laser", ("fragment_a", "fragment_b", "fragment_c"))
```

```python
molar._client.get_database_information()
```

```python
molar.get_all_fragments("fragment_a")
```

```python
for fragment_type in molar._fragments:
    descriptors = pd.read_csv(f"descriptors/gen2_pca_desc_{fragment_type[-1]}.csv")
    all_frags = molar.get_all_fragments(fragment_type)

    for idx, row in all_frags.iterrows():
        row = row.to_dict()
        frag_desc: list = descriptors[descriptors.hid==row["hid"]].iloc[0, 1:].values.astype(np.float).tolist()
        frag_desc_dict = {f"pca_desc_{i}": float(value) for i, value in enumerate(frag_desc)}
        molar._client.update_entry(
            uuid=row["molecule_id"],
            type="molecule",
            data={"descriptors": frag_desc_dict}
        )
```

```python
molar._client.query_database("synthesis", limit=100000)
```
